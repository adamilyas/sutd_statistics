```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ph_data = "112.9 112.4 116.2 125.1 132.3 129.9 127.2 124.0 123.4 122.7 125.0 126.0 \
126.1 125.5 123.5 123.1 122.6 122.9 120.5 125.3 127.7 124.2 121.3 117.7"
data = [float(i) for i in ph_data.split()]
```

```python
plt.plot(data)
plt.title("Q7: Time Series")
plt.xlabel("No. of months")
plt.ylabel("Gasoline Price/ Gallon (USD)")
plt.show()
```

```python
df = pd.DataFrame(data)
e1 = list(pd.ewma(df, alpha=0.3)[0])
e2 = list(pd.ewma(df, alpha=0.4)[0])
e3 = list(pd.ewma(df, alpha=0.5)[0])
```

```python
plt.plot(data, label="Original")
plt.plot(e1, label="alpha = 0.3")
plt.plot(e2, label="alpha = 0.4")
plt.plot(e3, label="alpha = 0.5")
plt.xlabel("No. of months")
plt.ylabel("Gasoline Price/ Gallon (USD)")
plt.title("Q7: Time Series")
plt.legend()
plt.show()
```

```python
import numpy as np

def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
print("MAPE for alpha = 0.3: " + str(mean_absolute_percentage_error(data[1:], e1[1:])))
print("MAPE for alpha = 0.4: " + str(mean_absolute_percentage_error(data[1:], e2[1:])))
print("MAPE for alpha = 0.5: " + str(mean_absolute_percentage_error(data[1:], e3[1:])))
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    T = len(y_true) #24
    summ = sum([np.abs((y_true[i] - y_pred[i])/y_true[i]) for i in range(1, T)])
    return round((1/(T-1))*summ*100, 3)
```

```python
print("MAPE for alpha = 0.3: " + str(mean_absolute_percentage_error(data, e1)))
print("MAPE for alpha = 0.4: " + str(mean_absolute_percentage_error(data, e2)))
print("MAPE for alpha = 0.5: " + str(mean_absolute_percentage_error(data, e3)))
```

```python
N = 4
numbers = [1, 2, 3, 4]
p = 0.25
def pop_variance(numbers):
    var = (sum([i**2 for i in numbers])/len(numbers)) - (sum(numbers)/len(numbers))**2
    return var
print("Population variance: " + str(pop_variance(numbers)))
```

```python
import itertools

def sample_variance(sample):
    n = len(sample)
    mean = sum(sample)/n
    var = 1/(n-1)*sum([(i-mean)**2 for i in sample])
    return var

for i in (itertools.combinations_with_replacement(numbers, 2)):
    print("Standard Deviation of " + str(i) + ": " + str(sample_variance(list(i))))
```
